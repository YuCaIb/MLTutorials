```python
from newspaper import Article, fulltext   # pip install newspaper3k 
import requests, string
```

```python
url = 'https://www.haber7.com/guncel/haber/3319295-erdoganin-dogal-gaz-mujdesinden-rahatsiz-oldular'
article = Article(url)
article.download()
article.parse()

text=article.text
```

```python
print(text)
```

```python
##### as far as this point we downloaded and convertet corpus to text with newspaper 
#### Now we will be cleaning our texts from stop words, commas and points then we will
#### consider and analyze our corpus with n-grams
```

```python
clean_text = text.translate(str.maketrans('','', string.punctuation)).lower()
print(clean_text)
## now, cleaned this corpus from punctuation
```

```python
### in order to analayz the corpus or do any kind of action we need to tokenize them.
### tokenizetion is breaking the text in to pieces mostly to words and here there are the other ways to do it: 
##  White space tokenization: Splitting the text into tokens by spaces, tabs, or newlines.
## Word tokenization: Splitting the text into tokens by words, using word boundaries or regular expressions.
## Sentence tokenization: Splitting the text into tokens by sentences, using punctuation or natural language processing techniques
## Subword tokenization: Splitting the text into tokens by breaking down words into smaller units, such as morphemes or character n-grams.
```

```python
### pip install nltk
## nltk is natural language toolkit and this helps us to find out stop words then we will be deleting them from out tokens
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
my_stopwords=set(stopwords.words('turkish'))
```

```python
##tokenization 
tokenized= nltk.word_tokenize(clean_text)
tokenized2=nltk.word_tokenize(text)
```

```python
print(tokenized)
```

```python
len(tokenized) ## wanted to check how many items do we have.
```

```python
print(tokenized2)
```

```python
len(tokenized2)-len(tokenized)
```

```python
##now will be clean our tokens from stopwords
words=[]
for token in tokenized:
    if token not in my_stopwords:
        words.append(token)

    print(words)
    
wordsnonclean=[]
for token in tokenized2:
    if token not in my_stopwords:
            wordsnonclean.append(token)
```

```python
len(words)
```

```python
### this is another function from nltk that provides to analyze frequencies of words in text,corpus,documents
from nltk import FreqDist
fdist = FreqDist(tokenized)
fdist2 = FreqDist(tokenized2)
print(fdist.most_common(50))
print(fdist2.most_common(50))
```

```python
### now we will be starting to analyze the text with n-grams it
## n-gram means that : to analayze a word what really means
## because meaning of a word mostly depends the words that comes before or after
```

```python
## pip install pandas // pandas helps to manupilate things
import pandas as pd
count_unigrams = (pd.Series(nltk.ngrams(words, 1)).value_counts())
count_unigramsnonclean = (pd.Series(nltk.ngrams(wordsnonclean, 1)).value_counts())



print(" corpus number of unigrams (more than 2):",len(count_unigrams[count_unigrams > 2]))
```

```python
print(" corpus number of unigrams (more than 2):",len(count_unigramsnonclean[count_unigramsnonclean > 2]))
print(count_unigramsnonclean)
## I wanted to see how many is the difference between non cleaned and cleaned from stopwords
```

```python
count_bigrams= (pd.Series(nltk.ngrams(words,2)).value_counts())
print(" corpus number of bigrams (more than 2):",len(count_bigrams[count_bigrams > 2]))
```

```python
count_trigrams = (pd.Series(nltk.ngrams(words, 3)).value_counts())
print("corpus number of trigrams (more than 2):",len(count_trigrams[count_trigrams > 2]))
```

```python
bigram_series= (count_bigrams) [:10] ## it means the first 10 elements
trigram_series = (count_trigrams) [:10]
print(bigram_series)
print(trigram_series)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')

from collections import Counter
def pos_taggin(tokens):
    tags=nltk.pos_tag(tokens)
    counts= Counter(tag for word, tag in tags)
    return counts
pos=pos_taggin(tokenized)
print(pos)
```
