```python
import nltk ## a library is really usefull for nlp building
##  https://wortschatz.uni-leipzig.de/en/download/Turkish 
## reading our database
f = open("tur_news_2005_10K-sentences.txt", "r", encoding='utf8')

```

```python
text=f.read()
f.close()
```

```python
clean_text=text.translate(str.maketrans('','',string.punctuation)).lower()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords_set= set(stopwords.words('turkish'))
```

```python
print(text)
```

```python
print(clean_text)
```

```python
## tokenazetion
tokens=nltk.word_tokenize(clean_text)

words=[]
for token in tokens:
    if token not in stopwords_set:
        words.append(token)
```

```python
## now we will be tagging every token with pos_tag() function for this we need to load tagger.
## it will help us to understand 
nltk.download('averaged_perceptron_tagger')
from collections import Counter
def nltk_postag(tokens):
    tags=nltk.pos_tag(tokens,lang='tur') 
    counts=Counter(tags)
    return counts

pos=nltk_postag(tokens)
print(pos)
```
