```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
with open ("harry.txt", "r") as f:
    text = f.read()
```

```python
print(text)
```

```python
##creating a doc container
doc= nlp(text) ## it tokineze the text
## paranthesis are tokeniezed as well
## take a look, don't is two token do is a token and n't (not) is another token
```
